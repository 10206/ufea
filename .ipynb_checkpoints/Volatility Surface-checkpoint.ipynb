```python
df = pd.read_csv('spx_quotedata.csv',skiprows = [1, 2])
df['Expiration Date'] = pd.to_datetime(df['Expiration Date'], format='%a %b %d %Y')
```

<h1> 우리가 필요한 Parameter = 함수에 들어갈 인수들 </h1>
<h3> S, K, T, r 그리고 Market price </h3>

```python
## 데이터 프레임 분리 ##
# Call, put의 가격은 다르고 S, K, T, r 은 다 공통이므로 이에 따라서 정리를 해줌.

#만기
df['Current'] = pd.to_datetime('2024-03-10')
df['Maturity'] = (df['Expiration Date'] - df['Current']).dt.days / 252

# 현재 주가. 3.10일 오전 9시 2분 데이터 이므로 그때의 주가를 가져옴
price_df = yf.download('^GSPC', period = "2mo")
stock_price = price_df.loc['2024-3-11']['Open']

### call option 데이터 프레임 ###

# OTM call을 뽑기위해 Strike가 S보다 큰 Call들로 정리
call_df = df[df['Strike'] >= stock_price]
call_df = call_df[call_df['Maturity'] > 0]
call_df = call_df[call_df["Volume"] > 0]
call_df = call_df[call_df["Bid"] > 0.05]
call_df = call_df[call_df["Ask"] > 0.05]
# bid-ask price를 mid price로 바꿈
call_df['midprice'] = (df['Ask']+df['Bid'])*0.5

call_df = call_df[['midprice', 'Strike', 'Maturity']]


### put option 데이터 프레임 ###

# OTM Put이므로 Strik가 S보다 작은 Put들을 추출
put_df = df[df['Strike'] <= stock_price]
put_df = put_df[put_df['Maturity'] > 0]
put_df = put_df[put_df["Volume"] > 0]
put_df = put_df[put_df["Bid"] > 0.05]
put_df = put_df[put_df["Ask"] > 0.05]
# mid price로 바꿈
put_df['midprice'] = (df['Ask.1']+df['Bid.1'])*0.5
# 필요한 Column만 추출
put_df = put_df[['midprice','Strike','Maturity']]

# 이자율, 3월 11일 기준 SOFR
rfr = 0.0531
q = 0 
```

```python
S = np.asarray(stock_price) # Underlying Price
cstrike = call_df.Strike.to_numpy() # Strike Price
pstrike = put_df.Strike.to_numpy() # Strike Price
maturity = call_df.Maturity.to_numpy() # Time to Maturity
pmaturity = put_df.Maturity.to_numpy()
r = np.asarray(rfr) # Interest Rate
q = np.asarray(q) # Dividend Rate

# 콜옵션 내재변동성 계산
call = np.asarray(1) # Option Type
call_price = call_df.midprice.to_numpy() # Market Price
civol = calcbsimpvol(dict(cp=call, P=call_price, S=S, K=cstrike, tau=maturity, r=r, q=q))

call_df['Imvol'] = civol

# 풋옵션 내재변동성 계산
put = np.asarray(-1) # Option Type
put_price = put_df.midprice.to_numpy() # Market Price
pivol = calcbsimpvol(dict(cp=put, P=put_price, S=S, K=pstrike, tau=pmaturity, r=r, q=q))

put_df['Imvol'] = pivol
```

```python
vdf = pd.concat([put_df,call_df])
vdf = vdf[['Strike','Maturity', "Imvol"]]
vdf = vdf[vdf['Imvol'].isnull() == False]
vdf = vdf[vdf['Imvol'] < 1]
vdf
```
